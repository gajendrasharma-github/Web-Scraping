```python
## Bottom Section

#About
brand2=[]
Series=[]
Colour2=[]
Form_factor=[]

# Dimensions
Item_height=[]
Item_Width=[]
Standing_screen_display_size=[]
Screen_Resolution=[]
Resolution=[]
Product_Dimensions=[]

Batteries=[]

# Processors 
Processor_brand=[]
Processor_type=[]
Processor_speed=[]
Processor_count=[]
RAM_Size=[]

# Memory
Memory_Technology=[]
Computer_Memory_Type=[]
Maximum_memory_supported=[]
Hard_disk_description=[]
Hard_drive_interface=[]
Speaker_Description=[]

# Graphics
Graphics_Coprocessor2=[]
Graphics_Chipset_Brand=[]
Graphics_card_description=[]
Graphics_card_ram_type=[]
Graphics_card_ram_size=[]
Graphics_card_interface=[]

#Connectivity
Connectivity_Type=[]
Wireless_type=[]
No_of_usb_ports=[]
No_of_hdmi_ports=[]

#Other Info
Wattage=[]
Optical_drive_type=[]
Operating_system=[]
Avg_Battery_Life=[]
Batteries_included=[]
No_of_Li_ion_cells=[]
Included_components=[]
Manufacturer=[]
Country_of_origin=[]
Item_weight=[]
```

```python
s=Service('C:/Users/sharm/Downloads/chromedriver-win32/chromedriver-win32/chromedriver.exe')

#Top Section
title=[]
rating=[]
No_of_ratings=[]
price=[]
mrp=[]

#Mid Section
brand=[]
Model_name=[]
Screen_size=[]
Colour=[]
Cpu_model=[]
RAM=[]
OS=[]
Special_feature=[]
Graphics_card=[]
Graphics_coprocessor=[]



for i in link_list:
    
    # Getting to URL via Selenium and Fetching the html page
    driver=webdriver.Chrome(service=s)
    url=i
    driver.get(url)
    webpage=driver.page_source
    
    
    # Using Beautiful Soup to do the thing in offline Mode
    soup=BeautifulSoup(webpage, 'html.parser')
    
    
    
    
    #Upper Section
    try:
        title.append(soup.find('span', attrs={'id':'productTitle', 
                                          'class':'a-size-large product-title-word-break'}).text.strip())
    except: 
        title.append(np.nan)
    
    try:
        rating.append(soup.find('a', attrs={'href':'javascript:void(0)', 'role':'button', 
                      'class':'a-popover-trigger a-declarative'}).find('span').text.strip())
    except:
        rating.append(np.nan)
    
    try:
        No_of_ratings.append(soup.find('span', attrs={'id':'acrCustomerReviewText',  
                      'class':'a-size-base'}).text.strip())
    except:
        No_of_ratings.append(np.nan)
    
    try:
        price.append(soup.find('span', attrs={'class':"a-price-whole"}).text.strip())
    
    except:
        price.append(np.nan)
        
    try:
        mrp.append(soup.find('span', attrs={'class':"a-price a-text-price"}).find('span').text.strip())
    except:
        mrp.append(np.nan)
        
    
    
    #Mid- Section
    
    try:
        brand.append(soup.find('table', attrs={'class':"a-normal a-spacing-micro"}).find('tr', attrs={'class':
        'a-spacing-small po-brand'}).find('span', attrs={'class':'a-size-base po-break-word'}).text.strip())
    except:
        brand.append(np.nan)
    
    try:
        Model_name.append(soup.find('table', attrs={'class':"a-normal a-spacing-micro"}).find('tr', attrs={'class':
        'a-spacing-small po-model_name'}).find('span', attrs={'class':'a-size-base po-break-word'}).text.strip())
    except:
        Model_name.append(np.nan)
        
    try:
        Screen_size.append(soup.find('table', attrs={'class':"a-normal a-spacing-micro"}).find('tr', attrs={'class':
            'a-spacing-small po-display.size'}).find('span', attrs={'class':'a-size-base po-break-word'}).text.strip())
    except:
        Screen_size.append(np.nan)
    
    try:
        Colour.append(soup.find('table', attrs={'class':"a-normal a-spacing-micro"}).find('tr', attrs={'class':
            'a-spacing-small po-color'}).find('span', attrs={'class':'a-size-base po-break-word'}).text.strip())
    except:
        Colour.append(np.nan)
    
    try:
        Cpu_model.append(soup.find('table', attrs={'class':"a-normal a-spacing-micro"}).find('tr', attrs={'class':
            'a-spacing-small po-cpu_model.family'}).find('span', attrs={'class':'a-size-base po-break-word'}).text.strip())
    except:
        Cpu_model.append(np.nan)
    
    try:
        RAM.append(soup.find('table', attrs={'class':"a-normal a-spacing-micro"}).find('tr', attrs={'class':
            'a-spacing-small po-ram_memory.installed_size'}).find('span', attrs={'class':'a-size-base po-break-word'}).text.strip())
    except:
        RAM.append(np.nan)
    
    try:
        OS.append(soup.find('table', attrs={'class':"a-normal a-spacing-micro"}).find('tr', attrs={'class':
            'a-spacing-small po-operating_system'}).find('span', attrs={'class':'a-size-base po-break-word'}).text.strip())
    except:
        OS.append(np.nan)
    
    try:
        Special_feature.append(soup.find('table', attrs={'class':"a-normal a-spacing-micro"}).find('tr', attrs={'class':
            'a-spacing-small po-special_feature'}).find('span', attrs={'class':'a-size-base po-break-word'}).text.strip())
    except: 
        Special_feature.append(np.nan)
    
    try:      
        Graphics_card.append(soup.find('table', attrs={'class':"a-normal a-spacing-micro"}).find('tr', attrs={'class':
            'a-spacing-small po-graphics_description'}).find('span', attrs={'class':'a-size-base po-break-word'}).text.strip())
    except:
        Graphics_card.append(np.nan)
    try:
        Graphics_coprocessor.append(soup.find('table', attrs={'class':"a-normal a-spacing-micro"}).find('tr', attrs={'class':
            'a-spacing-small po-graphics_coprocessor'}).find('span', attrs={'class':'a-size-base po-break-word'}).text.strip())
    except:
        Graphics_coprocessor.append(np.nan)
```
